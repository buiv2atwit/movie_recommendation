```python
#modify genres data
#['Action', 'Adventure', 'Fantasy', 'Science Fiction']
print(movies.iloc[0].genres)
print(type(movies.iloc[0].genres))
```

```python
#evaluate a string that contains a Python list
#the ast.literal_eval method successfully evaluates the string, and gets the Python list object.
import ast 
print(ast.literal_eval(movies.iloc[0].genres))
print(type(ast.literal_eval(movies.iloc[0].genres)))
```

```python
# function get all the name of each dictionary in the list
def convert(obj):
    L = [];
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
#implement convert funtion for every rows in column genres
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#implement convert funtion for every rows in column keywords
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#check the data in cast column
ast.literal_eval(movies.iloc[0].cast)
```

```python
# function get only 3 name of each dictionary in the list
def convertCast(obj):
    L = [];
    counter = 0;
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
        
    return L
```

```python
#implement convertCast funtion for every rows in column cast
movies['cast'] = movies['cast'].apply(convertCast)
```

```python
movies.head()
```

```python
#check the data in crew column
ast.literal_eval(movies.iloc[0].crew)
```

```python
# function get all the director of each dictionary in the list
def fetch_director(obj):
    L = [];
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break        
    return L
```

```python
# implement fetch_director for every rows in crew column
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# split every single word in overview for using to do recommended
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
# remove all the spaces in every rows in each columns
movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
# implement remove all the spaces for every columns
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#combine all the words in overview, genres, keywords, cast and crew columns 
#using tags columns for recommended system
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#create new dataFrame main_data for recommended system
main_data = movies[['movie_id', 'title', 'tags']]
```

```python
main_data.head()
```

```python
#join all the words in tags columns into big string
main_data['tags'] = main_data['tags'].apply(lambda x : " ".join(x))
```

```python
main_data.iloc[0].tags
```

```python
#convert all the words to lower case
main_data['tags'] = main_data['tags'].apply(lambda x : x.lower())
```
